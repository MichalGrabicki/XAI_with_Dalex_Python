```python
# Loading libraries
import pandas as pd
```

```python
# Loading data
filename = 'Poland_Macro_History.csv'
data_location = ''
data_raw = pd.read_csv(data_location + filename, sep=',')
```

```python
# Chcecking imported data 
data_raw.head()
```

```python
data_raw.tail()
```

```python
# Replacing space from column names with "_" and removing "-" 
data_raw.columns = data_raw.columns.str.replace(' ', '_')
data_raw.columns = data_raw.columns.str.replace('-', '')
```

```python
# Chcecking data structure
data_raw.info()
```

```python
# Changin date coloumn name and type
data_raw['Concept']= pd.to_datetime(data_raw['Concept'])
data_raw = data_raw.rename(columns={'Concept':'Date_key'})
```
